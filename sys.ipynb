### stdin, stdout, stderr (input, output, and error channels) control

```python
# sys.stdout : the source of print funtion that produce output
sys.stdout.write("hello")
```

```python
print("hello")
```

```python
# another usage of stdout to control the print channel
fd = open("my_stdout.txt","w")
sys.stdout = fd
print("hellllllllllllllo2")
# this will be printed in the my_stdout.txt not in the terminal
```

```python
# to redirect the stdout to the current directory
sys.stdout = sys.__stdout__
```

```python
print("hello")
```
